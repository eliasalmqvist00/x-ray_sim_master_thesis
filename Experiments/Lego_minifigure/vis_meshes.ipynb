## Imports

```python
import ipynbname
import pyvista as pv
from pyvistaqt import BackgroundPlotter

from pathlib import Path
import sys

path = ipynbname.path()
BASE = path.parents[2] 

sys.path.append(str(BASE))
path_data = BASE / "Data/Lego_minifigure"
```

## Visualise mesh function

```python
def visualise_mesh(mesh_paths, colors=["Gray","Blue","Red"], show_edges=True):
    
    meshes = [pv.read(path) for path in mesh_paths]

    #Make a window and show it
    plotter = BackgroundPlotter(show=True)
    for mesh, color in zip(meshes, colors):
        plotter.add_mesh(mesh, color=color, show_edges=show_edges)
    plotter.add_axes()
    plotter.show()
```

## Importing meshes

```python
path_STL_body = path_data / "Meshes/MeshBodyfastSimplified.stl"

path_STL_hands = path_data / "Meshes/Mesh from Hands fastsimplified.stl"

path_STL_print = path_data / "Meshes/Mesh from Print fastsimplified.stl"
```

## Adding the meshes to visualise to an array 

```python
meshes = [path_STL_body, path_STL_hands, path_STL_print]
```

## Visualising the mesh/meshes

```python
visualise_mesh(meshes, show_edges=False, colors=["light blue", "yellow", "light blue"])
```
